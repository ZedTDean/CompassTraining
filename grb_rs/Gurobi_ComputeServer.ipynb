```python

import gurobipy as gp
from gurobipy import GRB

env= gp.Env(empty=True)
env.setParam('CSMANAGER',"https://demos-manager.gurobi.com")
env.setParam('CSAPIACCESSID',"1e1f4a03-105c-4eef-85fa-6c6d40ec4f5f")
env.setParam('CSAPISECRET',"120ee9bd-a857-4244-8178-6a816acbacc1")
env.setParam('CSAPPNAME',"Databricks")
env.start()
```

```python
# This function formulates and solves the following MIP model (see mip1.py):
#  maximize
#        x +   y + 2 z
#  subject to
#        x + 2 y + 3 z <= 4
#        x +   y       >= 1
#        x, y, z binary
# Create a new model
m = gp.Model("mip1",env)
# Create variables
x = m.addVar(vtype=GRB.BINARY, name="x")
y = m.addVar(vtype=GRB.BINARY, name="y")
z = m.addVar(vtype=GRB.BINARY, name="z")

# Set objective
m.setObjective(x + y + 2 * z, GRB.MAXIMIZE)

# Add constraint: x + 2 y + 3 z <= 4
m.addConstr(x + 2 * y + 3 * z <= 4, "c0")

# Add constraint: x + y >= 1
m.addConstr(x + y >= 1, "c1")

# Optimize model
m.optimize()

for v in m.getVars():
  print('%s %g' % (v.varName, v.x))

print('Obj: %g' % m.objVal)
```

```python

m.dispose()
env.dispose()
```
